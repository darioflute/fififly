# Create latex files

```python
from fififly.flightplan import startDescription, writeLatex, runLatex
names = ['TRACY','TYLER','TENZIN','TUVOK','TATUM','TINA']
cycle = 'OC9T'

# Create description json file and extract figures from AORs
for name in names:
    aorfile = '202208_FI_' + name + '_SCI'
    # Create latex file
    writeLatex(aorfile, cycle, boresight=False)
    # Compile latex file and create pdf file
    runLatex(aorfile)
```
